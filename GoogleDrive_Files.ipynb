```python
# 07/30/2020-2
from __future__ import absolute_import
from IPython.display import clear_output
import glob, os, pathlib, sys
from os.path import *
from pathlib import Path
from google.colab import drive, files
homePath = Path.home()
cwdPath = Path.cwd()

def getGoogleDrive():
    try:
        drive.mount('gDrive', force_remount=True, timeout_ms=600000, 
            use_metadata_server=False)
    except Exception as err:
        print(err)

def addSysPath(path=None, quiet=True):
    sysPaths = sys.path.copy()
    counter = 0
    if not path==None and not path in sysPaths:
        sys.path.append(path)

    if not quiet:
        for pth in sysPaths:
            counter += 1
            if C: print(f'{C.WHITE}{counter}. {C.BLUE}{pth}')
            else: print(f'{counter}. {pth}')

addSysPath(quiet=False)
import FileUtilities
from FileUtilities import C
print(f'{C.GREEN}Ok')
```
